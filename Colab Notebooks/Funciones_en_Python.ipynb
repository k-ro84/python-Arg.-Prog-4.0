# Funciones

Una función es un mini programa dentro de un programa. Las funciones contienen varias sentencias bajo un solo nombre y constituyen una forma de dividir un proyecto grande en módulos pequeños.

Las funciones existen de modo autónomo con respecto al programa principal, cada una tiene su ámbito, sus variables locales. La división del código en funciones hace que las mismas se puedan reutilizar en su programa y en otros programas. Por lo tanto, podemos decir que una función es un subprograma que se llama dentro de un programa principal.

```
def funcion (parametros):
    bloque de codigo
```

## Ejemplo:

Escribir una función que nro es mayor

````python
def mayor (a , b):       #def: sirve para decirle a python que vamos a crear una funcion
    if a > b:
        print ('el mayor es ', a)
    else:
        print ('el mayor es ', b)
````

````python
#Programa principal
x = 10
y = 5
mayor (x , y)   #invocar a la funcion
````

## Parámetros de una función

````python
def potenciacion(base, exponente):
    return base ** exponente
````

````python
potenciacion(2, 5)
````

````python
potenciacion(5, 2)
````

````python
potenciacion(base = 3, exponente = 5)
````

````python
potenciacion(exponente = 5, base = 3)
````

````python
potenciacion(5)     #aca larga error ya que le falta un valor
````

## Parámetros con valores por defecto

````python
def potenciacion(base, exponente = 2):    #aca lo que hago es si exponente no recibe un valor para el argumento,va a valer por defecto 2
    return base ** exponente
````

````python
potenciacion(5, 3)    #aca el3 pisa la funcion de 2
````

````python
potenciacion(5)
````

## Funciones lambda

````python
def potenciacion(base, exponente):                         #se estructuran siempre en una unica linea de codigo
    return base ** exponente
````

````python
potenciacion_lambda = lambda base, exponente: base ** exponente        #los puntos separan el primer punto(los parametros de la funcion)
                                                                    #la segunda parte hace rferencia al resultado de la funcion(return)
 # el ejemplo de arriba con este se diferencia en que la funcion lambda me permite reducir linea de codigo
````

````python
potenciacion_lambda(3, 4)
````

## Comentarios en funciones

````python
def potenciacion(base, exponente):
    """
    funcion que calcula la potenciacion de una base a un exponente

    params:

    base: int
    exponente: int
    """

    return base ** exponente
````
